```python
# -------------------- TASK 2: API INTEGRATION --------------------

import requests   # library for API calls

print("\n--- FETCHING PRODUCTS ---")

# ---------- STEP 1: GET PRODUCTS ----------

url = "https://dummyjson.com/products?limit=20"

response = requests.get(url)   # sending GET request
data = response.json()         # converting response to JSON

products = data["products"]    # list of products

# printing table header
print("\nID | Title | Category | Price | Rating")
print("-" * 60)

# printing product details
for p in products:
    print(p["id"], "|", p["title"], "|", p["category"], "| $", p["price"], "|", p["rating"])


# ---------- STEP 2: FILTER + SORT ----------

print("\n--- FILTERED (rating >= 4.5) ---")

filtered = []

# filter products
for p in products:
    if p["rating"] >= 4.5:
        filtered.append(p)

# sort by price (descending)
filtered.sort(key=lambda x: x["price"], reverse=True)

# print filtered list
for p in filtered:
    print(p["title"], "- $", p["price"], "- Rating:", p["rating"])


# ---------- STEP 3: CATEGORY SEARCH ----------

print("\n--- LAPTOP PRODUCTS ---")

url2 = "https://dummyjson.com/products/category/laptops"

response2 = requests.get(url2)
data2 = response2.json()

laptops = data2["products"]

for p in laptops:
    print(p["title"], "- $", p["price"])


# ---------- STEP 4: POST REQUEST ----------

print("\n--- ADD PRODUCT (POST) ---")

new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

response3 = requests.post(
    "https://dummyjson.com/products/add",
    json=new_product
)

result = response3.json()

print("\nResponse from server:")
print(result)
```

```python
# -------------------- TASK 3: EXCEPTION HANDLING --------------------


# Part A: Safe Division Function

def safe_divide(a, b):
    try:
        result = a / b   # try normal division
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


# Testing the function
print(safe_divide(10, 2))     # normal case
print(safe_divide(10, 0))     # division by zero
print(safe_divide("ten", 2))  # wrong type


# Part B: Safe File Reading

def read_file_safe(filename):
    try:
        file = open(filename, "r")   # try opening file
        content = file.read()        # read full content
        file.close()
        return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        # this will always run (important point)
        print("File operation attempt complete.")


# Testing
print(read_file_safe("python_notes.txt"))   # should work (if file exists)
print(read_file_safe("ghost_file.txt"))     # will show error

# Part C: Robust API Calls

import requests

# Function to get products safely
def get_products():
    url = "https://dummyjson.com/products?limit=5"

    try:
        response = requests.get(url, timeout=5)
        data = response.json()

        # printing some product titles
        for product in data["products"]:
            print(product["title"])

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


# calling function
get_products()


# Part D: Input Validation Loop

import requests

while True:
    user_input = input("Enter a product ID to look up (1–100), or 'quit' to exit: ")

    # exit condition
    if user_input.lower() == "quit":
        print("Exiting program...")
        break

    # validation check
    if not user_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    product_id = int(user_input)

    if product_id < 1 or product_id > 100:
        print("Please enter a number between 1 and 100.")
        continue

    # API call
    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)

        if response.status_code == 404:
            print("Product not found.")
        elif response.status_code == 200:
            data = response.json()
            print("Title:", data["title"])
            print("Price:", data["price"])
        else:
            print("Something went wrong.")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)



```

```python
 # -------------------- TASK 4: LOGGING TO FILE  --------------------

import requests
from datetime import datetime


# 🔹 function to log errors
def log_error(function_name, error_type, message):
    time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log_line = f"[{time}] ERROR in {function_name}: {error_type} — {message}\n"

    with open("error_log.txt", "a") as f:
        f.write(log_line)


# 🔹 1. Connection Error Test (fake URL)
try:
    print("Testing connection error...")
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)

except requests.exceptions.ConnectionError:
    print("Connection error triggered ✅")
    log_error("fetch_products", "ConnectionError", "No connection could be made")

except Exception as e:
    log_error("fetch_products", "Exception", str(e))


# 🔹 2. HTTP 404 Test
try:
    print("Testing HTTP error...")

    product_id = 999
    url = f"https://dummyjson.com/products/{product_id}"

    response = requests.get(url, timeout=5)

    if response.status_code != 200:
        print("HTTP error triggered ✅")
        log_error(
            "lookup_product",
            "HTTPError",
            f"{response.status_code} Not Found for product ID {product_id}"
        )

except Exception as e:
    log_error("lookup_product", "Exception", str(e))


# 🔹 3. Print log file
print("\n--- Log File Content ---")

try:
    with open("error_log.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("Log file not found.")
```
